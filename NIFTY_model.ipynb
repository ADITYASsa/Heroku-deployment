# MODELLING OF DATASET

```python
X = data.loc[:,data.columns!='close']
X
```

```python
y=data['close']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=0)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_train_std=sc.fit_transform(X_train)
```

```python
X_test_std=sc.transform(X_test)
```

```python
X_train_std
```

# svm linear

```python
from sklearn.svm import SVR
```

```python
dataframelinear = SVR(kernel='linear', C = 22)
```

```python
dataframelinear.fit(X_train_std,y_train)
```

```python
y_train_pred = dataframelinear.predict(X_train_std)
```

```python
y_test_pred = dataframelinear.predict(X_test_std)
```

```python
result = pd.DataFrame({'Actual' : y_test , 'Predicted' : y_test_pred}).reset_index()[['Actual' , 'Predicted']]
result
```

```python

plt.plot(y_test,result['Predicted'],color='red')
plt.plot(y_test,result['Actual'],color='yellow')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test,y_test_pred)
print("Mean_squared_error is: ", mse)
```

```python
r2score = r2_score(y_test , y_test_pred)
print("R2 score is: " , r2score)
```
